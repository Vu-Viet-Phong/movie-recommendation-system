### Merge datasets

```python
dataset = pd.merge(pd.merge(movies, ratings), users)
```

```python
dataset.head()
```

```python
dataset['Age'].replace(1, 'Under 18', inplace=True)
dataset['Age'].replace(18, '18-24', inplace=True)
dataset['Age'].replace(25, '25-34', inplace=True)
dataset['Age'].replace(35, '35-44', inplace=True)
dataset['Age'].replace(45, '45-49', inplace=True)
dataset['Age'].replace(50, '50-55', inplace=True)
dataset['Age'].replace(56, '56+', inplace=True)
```
